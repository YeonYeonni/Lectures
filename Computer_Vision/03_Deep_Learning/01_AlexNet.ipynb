## Image load and normalize

```python
image = './data/house.jpg'
image = Image.open(image).convert('RGB')
plt.imshow(image)
plt.show()
```

- normalize : RGB scale을 통일

```python
# convert tensor and normalize image
to_tensor = torchvision.transforms.ToTensor()

# Imagenet에 대한 normalize
normalizer = torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                               std=[0.229, 0.224, 0.225])

image = normalizer(to_tensor(image))
image = image.unsqueeze(0) # batchify [B, C, H, W]
print(image.shape)
```

## Load pre-trained weights on ImageNet classification task

```python
alexnet = models.alexnet(pretrained=True)
alexnet.eval() # freeze weight
alexnet.train() # trainable weight
alexnet.eval()
```

## Forward

```python
print("input shape: ", image.shape)
logit = alexnet(image)
print("output shape", logit.shape, logit)
print(torch.argmax(logit)) # 660번째 class라고 예측
```

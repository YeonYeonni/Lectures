### (1) 단일 변수 함수

**테일러 급수(Taylor series) 전개 : 함수 f(x)를 한 점 x = a에 대해 테일러 급수 전개하면 무한급수가 됨**


**매클로린 급수(MacLaurin series) : a = 0인 경우의 테일러 급수**

테일러 급수의 특징

- 테일러 급수 전개는 유일하다
- 어떤 두 함수도 동일한 테일러 급수가 되지 않음
- x = a에서 f(x)의 모든 미분이 존재하고 급수가 수렴할 때 의미가 있음
- x가 충분히 a에 가까울 때, 즉 |x - a| <= ε일 때 수렴
    - ε는 수렴반경
    - 많은 경우 ε은 무한

테일러 급수의 임의값 x에 대한 전개


처음의 n + 1 항만 채택
- 무한급수의 항으로 모든 값을 계산하기는 불가능
- 식에서 급수를 절단하는 효과는 큰 중요성을 가짐


En은 절단오차 (절단된 항의 합)
- 절단오차의 범위


ξ는 범위 (x: x + h) 안의 어떤 값
- En에 대한 식은 급수의 첫 번째 폐기항과 동일. x를 ξ로 치환한 것
- ξ의 값이 불확정이므로 위 En에 대한 식에서 알 수 있는 것은 절단오차의 상한과 하한 범위
- 따라서 En을 아래와 같이 축약
    - 절단오차가 h^{n+1}의 차수 또는 h^{n+1}로 거동한다고 하는 간략한 방법
    - 만약 h가 수렴반경 안에 있으면 (어떤 항이 절단오차에 추가되면) 오차는 항상 축소됨


**중앙값 정리 (mean value theorem) : 특별한 경우로 n = 1일 때의 테일러 정리**

- 연립방정식을 다루지 않고 수치해석을 수행하기는 거의 불가능
- 연립방정식의 규모가 매우 크기 때문에 많은 계산 자원을 소모

**따라서 대규모 계수행렬을 특별한 형태로 만들어 풀고자 하는 알고리즘 개발**

### (1) 공학 분야의 선형연립방정식

1. 시스템이 이산적이면, 체계는 직접적으로 선형방정식
- 수치해석은 이산변수만을 다룰 수 있음
- 미분방정식을 연립대수방정식으로 근사

2. 시스템이 연속적이면, 대수방정식이 아닌 미분방정식


**선형연립방정식의 모형화는 Ax = b 형태의 방정식**
- x는 입력, b는 반응 (출력)
- 계수행렬 A는 시스템의 특성. 입력과 독립.

### (2) 선형연립방정식의 표기

A11x1 + A12x2 + ... + A1nxn = b1
A21x1 + A22x2 + ... + A2nxn = b2
...
...
An1x1 + An2x2 + ... + Annxn = bn

- 계수 Aij와 우변의 bj은 기지값
- xi는 미지값

**행렬로 표기**


**간단하게 표기**
- Ax = b

**확대계수행렬 (augmented coefficient matrix) 또는 확대행렬 (augmented matrix)**
- 연립방정식을 계산 목적으로 나타낼 때 유용한 형태
- 계수행렬 A의 오른쪽에 상수벡터 b를 추가한 형태



### (3) 해의 유일성

n개의 미지수와 n개의 선형방정식으로 된 연립방정식
- 계수행렬의 **행렬식 (determinant)** 이 **비특이 (nonsingular)** (즉, |A| /= 0 라면), 유일한 해를 가짐
- 비특이 행렬의 행과 열은 **선형 독립 (linearly independent)**
- 어떤 행(또는 열)도 다른 행(또는 열)의 선형 조합이 되지 않음

**계수행렬이 특이 (singular) 라면?**
- 연립방정식은 무한한 수의 해를 갖거나 해를 가지지 않음
    - ex1) 첫번째 방정식이 두번째 방정식의 상수배이면 x와 y의 쌍은 무엇이든지 둘째 방정식의 해가 됨
        - 이런 상태를 **선형 종속 (linearly dependent)** 이라고 함
    
    - ex2) 첫번째 방정식과 두번째 방정식이 모순
        - 해가 없음

### (4) 호조건과 악조건

**계수행렬이 거의 특이 (행렬식 |A|이 매우 작을 경우) 하다면?**
- 이를 판별하기 위해 행렬식의 크기에 대한 기준이 필요
- 이 기준을 **행렬 노옴 (norm)** 이라 부르며, ||A||로 표기
- |A| << ||A||인 경우 행렬식이 작음

**유클리드 노옴 (Euclide norm)**


**행합 노옴 (row-sum norm) 또는 무한 노옴 (infinity norm)**


**행렬조건수 (matrix condition number)**


- 악조건의 형식적 척도
- 행렬조건수가 1에 가까우면 **호조건 (well-conditioned)**
- 조건수는 악조건 (ill-conditioned)의 정도에 따라 커지며 특이행렬에서는 무한대에 이름
- 조건수는 행렬노옴의 선택에 따라 달라짐. 유일하지 않음
- 대규모 행렬의 조건수를 계산하기는 어려움 -> 행렬식과 행렬요소의 크기를 비교해 조건수를 가늠

**방정식이 악조건이면, 계수행렬에서 작은 변화가 생겨도 해에는 큰 변화를 초래**
- 행렬식이 매우 작으며, 조건수가 매우 크다면 방정식은 악조건
    - ex) y의 계수를 0.1% 바꾸었을때, 해의 변화가 100%
- 악조건 방정식의 수치해는 신뢰할 수 없음
    - 마무리오차가 계수 행렬에서 계수의 작은 변화를 만드는 것과 같은 효과를 불러옴
    - 악조건의 심각성에 따라 오차의 크기가 달라짐

### (5) 해석 방법

**직접법과 반복법**
1. 직접법
- 원 방정식을 손쉽게 풀 수 있는 등가방정식 (같은 해를 갖는 방정식) 으로 변형
- 세 가지 조작을 적용
    - 1) 두 방정식의 교환 (|A|의 부호의 변경)
    - 2) 0이 아닌 상수를 방정식에 곱하기 (같은 상수를 |A|에 곱하기)
    - 3) 0이 아닌 상수를 한 방정식에 곱하고 다른 방정식에서 이 방정식을 빼기 (|A|는 변화되지 않음)
    
2. 반복법 (간접법)
- 해 x를 가정. 어떤 수렴 조건에 이를 때까지 이 해를 반복적으로 개선
- 일반적으로 직접법보다 비효율적
- 계수행렬이 매우 크고 성기다면 (대부분의 계수가 0) 계산적으로 큰 장점

### (6) 직접법 둘러보기

**세 가지 직접법**
- U는 상삼각행렬, L은 하삼각행렬, I는 항등행렬

1. Gauss 소거법
- 초기 형태 : Ax = b
- 최종 형태 : Ux = c

2. LU 분해
- 초기 형태 : Ax = b
- 최종 형태 : LUx = b

3. Gauss-Jordan 소거법
- 초기 형태 : Ax = b
- 최종 형태 : Ix = c

**삼각 행렬 (triangular matrix) : 정방행렬 A가 주대각의 어느 한쪽이 0으로만 되어 있는 행렬**
- U는 상삼각행렬. L은 하삼각행렬
- 삼각행렬은 많은 경우 계산을 매우 간단하게 함

**전방 대입 (forward substitution) : 처음 방정식에서 시작해 전진 방향으로 방정식을 풀어가는 것**

**후방 대입 (back substitution) : 마지막 방정식에서 시작해 후진 방향으로 방정식을 풀어가는 것**

### (7) 반복법 둘러보기

- 해 x의 초기값을 가정하고 시작하여, x의 변화가 무시할 수 있을 정도가 될 때까지 해를 반복적으로 갱신
- 반복법은 직접법보다 느림

**반복법의 장점**
- 1. 많은 문제에서 계수행렬을 전혀 저장할 필요가 없다
    - 계수행렬에서 0이 아닌 요소만 저장할 수 있다
    - 성긴 매우 큰 행렬을 다룰 수 있다
- 2. 자기 보정이 가능하다
    - 한 반복 주기에서 반올림오차를 다음 주기에서 보정할 수 있다
    
**반복법의 단점**
- 계산된 값이 항상 해에 수렴한다는 보장이 없다
